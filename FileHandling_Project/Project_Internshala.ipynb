```python
# Importing the libraries

import pandas as pd
import re
from collections import defaultdict
import string
```

```python
# Reading the file

df = pd.read_csv('C:/Users/USER/Desktop/FileHandling_Project/conv.txt' ,delimiter = ':',header = None)
```

```python
df.columns= ['Speaker','Dialogue']
```

```python
df.head()
```

```python
speaker_list = df['Speaker']
speaker_list
```

```python
# list of all unique speakers

all_speakers = [*set(df['Speaker'])]
all_speakers
```

```python
dialogue_list = list(df['Dialogue'])
dialogue_list
```

```python
# Converting all the dialogues into lower case

dialogue_lower =[]

for i in dialogue_list:
    i_lower = i.lower()
    dialogue_lower.append(i_lower)
    
dialogue_lower
```

```python
# Removing all the special characters

special_removal= []

for j in dialogue_lower:
    special_removal.append(re.sub("[.,:'?!…]","",j))
    
```

```python
special_removal
```

```python
# creating new column afer removing special characters and converting into lower case

df['sc_removed_dialogue']= special_removal
```

```python
df
```

```python
# Making the speaker column as the index

df.set_index('Speaker',inplace=True, drop=True)
```

```python
df.head()
```

```python
A = sorted(set(df.loc['JON']['sc_removed_dialogue'].str.split().sum()))

unique_words = []

for i in A:
    for j in i:
        if j in string.punctuation:
            i = i.replace(j,"")   
    unique_words.append(i)

# creating a new file with character name and storing it with all unique words
with open('Project/JON.txt', 'w', encoding="utf_8",) as f:
    for line in unique_words:
        f.write(f"{line}\n")
```
